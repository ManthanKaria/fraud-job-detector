```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Load data
df = pd.read_csv("../data/fake_job_postings.csv")
df = df[['description', 'fraudulent']].dropna()
```

```python
# Class balance
df['fraudulent'].value_counts().plot(kind='bar', title='Fraudulent vs Real Jobs')
plt.show()
```

```python
# WordCloud for fraudulent job descriptions
fraud_text = " ".join(df[df['fraudulent']==1]['description'].tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(fraud_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
